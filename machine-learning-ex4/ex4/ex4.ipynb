# Part 3: Compute Cost (Feedforward)

```python
lambda_val = 0;
```

```python
def sigmoid(z):
    
    g = 1.0/(1 + np.exp(-z)  )
    
    return g 
```

```python
def nnCostFunction(nn_params, input_layer_size, hidden_layer_size, num_labels, x, y, lambda_val):
  
    Theta1 = ( np.reshape(nn_params[:(input_layer_size+1)*hidden_layer_size], [ hidden_layer_size,input_layer_size+1,] ) )
    Theta2 = ( np.reshape(nn_params[(input_layer_size+1)*hidden_layer_size:], [ num_labels,hidden_layer_size+1] ) )

    m = np.shape(x)[0]
    
    J = 0;
    Theta1_grad = np.zeros_like(Theta1)
    Theta2_grad = np.zeros_like(Theta2)
    
    Y = np.zeros([m,num_labels] )
    
    for i in range(m):
        Y[i,y[i]-1] = 1
    
    a1 = (np.hstack((np.ones([m,1]), x)))    

    a2 = sigmoid(np.dot(a1,np.transpose(Theta1)))
    a2 = np.hstack((np.ones([np.shape(a2)[0],1]), a2))
    
    a3 = sigmoid(np.dot(a2,np.transpose(Theta2)))

    J = - 1.0/m * np.sum( (Y * np.log( a3 ) + (1 - Y ) * np.log(1 - a3 )  ) ) \
    + lambda_val / ( 2 * m) * ( np.sum( Theta1[:,1:]**2 ) + np.sum( Theta2[:,1:]**2) )       
   
    # Second part
    
    delta3 = a3 - Y 
    
    delta2 = ( np.dot( delta3 ,  Theta2) * a2 * (1 - a2) )[:, 1:] 
    delta1 = ( np.dot( delta2 ,  Theta1) * a1 * (1 - a1) )[:, 1:] 

    Delta2 = np.dot( np.transpose(delta3) , a2)
    Delta1 = np.dot( np.transpose(delta2) , a1) 
    
    Theta1_grad = 1/m * Delta1
    Theta1_grad[:,1:] = Theta1_grad[:,1:] + lambda_val /m * Theta1[:,1:]
    
    Theta2_grad = 1/m * Delta2
    Theta2_grad[:,1:] = Theta2_grad[:,1:] + lambda_val /m * Theta2[:,1:]
     
    grad =  np.hstack((Theta1_grad.flatten(), Theta2_grad.flatten()))
    
    return J, grad
```

```python
def nnCostFunction_simpler(nn_params, input_layer_size, hidden_layer_size, num_labels, x, y, lambda_val):
  
    Theta1 = ( np.reshape(nn_params[:(input_layer_size+1)*hidden_layer_size], [ hidden_layer_size,input_layer_size+1,] ) )
    Theta2 = ( np.reshape(nn_params[(input_layer_size+1)*hidden_layer_size:], [ num_labels,hidden_layer_size+1] ) )

    m = np.shape(x)[0]
    
    Theta1_grad = np.zeros_like(Theta1)
    Theta2_grad = np.zeros_like(Theta2)
    
    Y = np.zeros([m,num_labels] )
    
    for i in range(m):
        Y[i,y[i]-1] = 1
    
    #a1 = np.transpose(np.hstack((np.ones([m,1]), x)))
    a1 = (np.hstack((np.ones([m,1]), x)))    

    a2 = sigmoid(np.dot(a1,np.transpose(Theta1)))
    a2 = np.hstack((np.ones([np.shape(a2)[0],1]), a2))
    
    a3 = sigmoid(np.dot(a2,np.transpose(Theta2)))

    J = - 1.0/m * np.sum( (Y * np.log( a3 ) + (1 - Y ) * np.log(1 - a3 )  ) ) \
    + lambda_val / ( 2 * m) * ( np.sum( Theta1[:,1:]**2 ) + np.sum( Theta2[:,1:]**2) )       
    
    return J
```

# Part 7: Implement Backpropagation 

```python
def debugInitializeWeights(fan_out, fan_in):
    
    W = np.transpose(np.reshape(np.sin( np.arange(fan_out * (1 + fan_in)) + 1 ), [ fan_in + 1 , fan_out])) / 10
 
    return W
    
```

```python
def computeNumericalGradient(J, theta, input_layer_size, hidden_layer_size, num_labels, X, y, lambda_val):
    
    numgrad = np.zeros(np.shape(theta)[0]);
    perturb = np.zeros(np.shape(theta)[0]);
    
    e = 1e-4

    for p in range(np.shape(theta)[0]):
        
        perturb[p] = e

        loss1 = J(theta - perturb, input_layer_size, hidden_layer_size, num_labels, X, y, lambda_val )[0]
        loss2 = J(theta + perturb, input_layer_size, hidden_layer_size, num_labels, X, y, lambda_val )[0]

        numgrad[p] = (loss2 - loss1) / (2 * e)
        perturb[p] = 0
    
    return numgrad
```

```python
def checkNNGradients( lambda_val = 0 ):
    
    input_layer_size = 3
    hidden_layer_size = 5
    num_labels = 3
    m = 5
    
    Theta1 = debugInitializeWeights(hidden_layer_size, input_layer_size  )
    Theta2 = debugInitializeWeights(num_labels, hidden_layer_size )

    X  = debugInitializeWeights(m, input_layer_size - 1);
    y  = 1 + np.mod(np.arange(m)+1, num_labels)
    
    nn_params = np.hstack((Theta1.flatten(), Theta2.flatten()))
    
    cost, grad = nnCostFunction(nn_params, input_layer_size, hidden_layer_size, num_labels, X, y, lambda_val)
    
    numgrad = computeNumericalGradient(J = nnCostFunction, theta = nn_params, input_layer_size = input_layer_size, \
                                       hidden_layer_size = hidden_layer_size , \
                                       num_labels = num_labels, X = X, y = y, \
                                       lambda_val = lambda_val)
    
    for i in range(np.shape(nn_params)[0]):
        print(f'{grad[i]:.6f} {numgrad[i]:.6f}' )
    
    print('\nThe above two columns you get should be very similar.' )
    print('(Left-Your Numerical Gradient, Right-Analytical Gradient)')
    
    diff = np.linalg.norm(numgrad-grad)/np.linalg.norm(numgrad+grad)
    
    print('\nIf your backpropagation implementation is correct, then')
    print('the relative difference will be small (less than 1e-9).')
    print(f'\nRelative Difference: {diff}');
    
```

# Part 8: Training NN

```python
from scipy.optimize import minimize
```

```python
def nnCostFunction2(nn_params, Xy):
    
    return nnCostFunction_simpler(nn_params, Xy[0], Xy[1], Xy[2], Xy[3], Xy[4], Xy[5])

def  nnCostFunction2_jac(nn_params, Xy):
    
    return nnCostFunction(nn_params, Xy[0], Xy[1], Xy[2], Xy[3], Xy[4], Xy[5])[1]
```

```python
def predict(Theta1, Theta2, x):
    
    m = np.shape(x)[0]
    
    num_labels = np.shape(Theta2)[0]

    p = np.zeros([m])
    
    a1 = np.ones([m,np.shape(x)[1]+1])
    a1[:,1:] = x
    
    z2 = np.dot(a1,np.transpose(Theta1))
    a2_prime = sigmoid( z2 )
    
    a2 = np.ones([m,np.shape(a2_prime)[1]+1])
    a2[:,1:] = a2_prime 
    
    z3 = np.dot(a2,np.transpose(Theta2))
    a3 = sigmoid( z3 )
    
    p = np.argmax( a3 , axis = 1) +1 
    
    return p
```

```python
lambda_val = 1
```

```python
constrains_args = [input_layer_size, hidden_layer_size, num_labels, X, y, lambda_val]
```

```python
%%time 
res = minimize(nnCostFunction2, x0 = nn_params, args=constrains_args, options ={'disp':True, 'maxiter':50 }, method="Newton-CG", jac = nnCostFunction2_jac )
```

```python
Theta1 = ( np.reshape(res.x[:(input_layer_size+1)*hidden_layer_size], [ hidden_layer_size,input_layer_size+1,] ) )
Theta2 = ( np.reshape(res.x[(input_layer_size+1)*hidden_layer_size:], [ num_labels,hidden_layer_size+1] ) )
```
