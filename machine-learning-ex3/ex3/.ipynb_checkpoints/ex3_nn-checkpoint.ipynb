# Part 2a: Vectorize Logistic Regression

```python
theta_t = np.array([-2, -1, 1, 2])
theta_t
```

```python
X_t = np.hstack( (np.ones([5,1]), np.reshape(np.arange(15)+1,[5,3], order='F' )/10))
y_t = np.array(np.array([1,0,1,0,1]) >= 0.5, dtype=np.int)
```

```python
lambda_t = 3
```

```python
def sigmoid(z):
    
    g = 1.0/(1 + np.exp(-z)  )
    
    return g 
```

```python
def lrFunctionReg(theta, X, y, lambda_val):

    m = np.shape(y)[0]
    
    J = np.sum(-y *np.log( sigmoid( np.dot( X, theta ) ) ) - ( 1 - y )*np.log( 1 - sigmoid( np.dot( X,theta ) ) ))/m + lambda_val/(2*m)*np.dot(theta[1:], theta[1:])
    
    grad = np.dot((sigmoid( np.dot(X,theta) ) -y  ),X )/m
    
    grad[1:] = grad[1:] + lambda_val / (m) * theta[1:]
    
    return J, grad
```

```python
J, grad = lrFunctionReg(theta_t, X_t, y_t, lambda_t )
```

```python
print(f'Cost: {J}');
print('Expected cost: 2.534819');
print('Gradients:');
print(f' {grad}');
print('Expected gradients:');
print(' 0.146561  -0.548558  0.724722 1.398003 ');
```

#  Part 2b: One-vs-All Training

```python
lambda_val = 0.1
```

```python
def lrFunctionReg2(theta, Xy):
    
    return lrFunctionReg(theta, Xy[0], Xy[1], Xy[2] )[0]
```

```python
theta_t
```

```python
from scipy.optimize import minimize

def oneVsAll(x, y, num_labels, lambda_val):
    
    m, n =  np.shape(x)
    all_theta = np.zeros([num_labels, n+1])
    X = np.ones([m,n+1])
    X[:,1:] = x
    
    initial_theta = np.zeros([n + 1])
    for c in range(1, num_labels+1):
        
        print(c)
        constrains_args = [X, np.array(y== c, dtype=np.int), lambda_val]
    
        #print(lrFunctionReg(initial_theta, constrains_args[0], constrains_args[1], constrains_args[2] ))
        
        res = minimize(lrFunctionReg2, x0 = initial_theta, args=constrains_args, options ={'disp':True}, method='SLSQP' )
        all_theta[ c-1 ] =  res.x
          
    return all_theta
    
```

```python
all_theta = oneVsAll(X, y, num_labels, lambda_val)
```

```python
def predictOneVsAll(all_theta, x):
    
    m, n =  np.shape(x)
    
    X = np.ones([m,n+1])
    X[:,1:] = x
    
    predict = np.zeros([m]) 
    num_labels = np.arange(np.shape(all_theta)[0])  
    
    for c in num_labels:

        predict[  np.dot( X,  all_theta[c]   )  >= 0.5  ]  = c+1 
        
    return predict
    
```

```python
predict = predictOneVsAll(all_theta, X)
```

```python
print(f'Training Set Accuracy: { np.mean( predict == y) * 100}%' )
```
